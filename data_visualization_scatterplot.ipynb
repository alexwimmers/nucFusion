```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating the dataframes

df1 = pd.read_excel("fusion_total_v08_awi_28-03-2024.xlsx", sheet_name = "LCOE_CALC_PLOT")
df2 = pd.read_excel("fusion_total_v08_awi_28-03-2024.xlsx", sheet_name = "LCOE_REF_PLOT")

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the first dataframe as a scatterplot with different colors for each "attribut"
sns.scatterplot(data=df1, x='OCC 2019', y='Wert', hue='Attribut', palette='viridis', ax=ax1)

# Create a second y-axis sharing the same x-axis
#ax2 = ax1.twiny()

# Plot the second dataframe as a line plot with different line types for each "references"
sns.lineplot(data=df2, y='Cost', x=df1["OCC 2019"],  hue = "Reference", palette='husl', linestyle='-', ax = ax1)

#ax2.set_xlim(df1['OCC 2019'].min(), df1['OCC 2019'].max())

# Set labels and legend
ax1.set_xlabel('OCC')
ax1.set_ylabel('Wert')
#ax2.set_xlabel('Cost')
ax1.legend(loc='upper right')
ax2.legend()


plt.title('Combined Plot of Dataframe 1 and Dataframe 2')

plt.tight_layout()
plt.show()
```
